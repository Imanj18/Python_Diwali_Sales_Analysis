```python
#STATE
```

```python
Orders_State=df.groupby("State")[['Orders']].sum().sort_values(by='Orders',ascending=False).head(5)
ax=Orders_State.plot(kind="bar")
for container in ax.containers:
    ax.bar_label(container, fmt='%d')
plt.title("Orders According to State")
plt.ylabel("Count of Orders")
#orders data on the basis of state
```

```python
Sales_State=df.groupby("State")[['Amount']].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='State', y='Amount',data=Sales_State, palette='Reds',hue='State')
sns.set(rc={'figure.figsize':(30,8)})
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d')
plt.title("Sales According to State")
plt.ylabel("Count of Orders")
#sales data on basis of state 
```

```python

# Group by Product_ID and Gender, then count orders
product_counts = df.groupby(["Product_ID", "Gender"]).size().reset_index(name="Order Count")

# Get the top 5 most and least ordered products across both genders
most_ordered = product_counts.nlargest(5, "Order Count")["Product_ID"]
least_ordered = product_counts.nsmallest(5, "Order Count")["Product_ID"]

# Combine both most and least ordered products using pd.Index
selected_products = pd.Index(most_ordered).append(pd.Index(least_ordered))

# Filter dataset to only show selected products
filtered_df = product_counts[product_counts["Product_ID"].isin(selected_products)]

# Plot the comparison graph
plt.figure(figsize=(10, 5))
sns.barplot(data=filtered_df, x="Product_ID", y="Order Count", hue="Gender", palette=["blue", "pink"])
plt.xlabel("Product ID")
plt.ylabel("Order Count")
plt.title("Comparison of Most and Least Ordered Products by Gender")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```
